```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/Shreyas3108/Weather/master/weather.csv')
```

```python
print(df)
```

```python
plt.plot(df['MinTemp'],df['MaxTemp'])
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.show()
```

```python
plt.scatter(df['MinTemp'],df['MaxTemp'])
```

```python
plt.plot(df['MinTemp'])
```

```python
plt.hist(df['MinTemp'])
```

```python
sns.distplot(df['Rainfall'],kde='True')
```

```python
sns.distplot(df['MaxTemp'],kde=True,rug=True,hist=False)
```

```python
sns.kdeplot(df['MaxTemp'],shade=True,color="r")
```

```python
sns.kdeplot(df['MaxTemp'],shade=True,color="r",bw=1000)
```

```python
plt.figure(figsize=(10,8))
sns.kdeplot(df['MaxTemp'],df['MinTemp'])
```

```python
plt.figure(figsize=(10,8))
sns.kdeplot(df['MaxTemp'],df['MinTemp'],shade=True)
```
